```python
import pandas as pd
import numpy as np
```

```python
def numeros_aleatorios(num):
  nums= [num]
  nums2= []
  m_r = []
  val_1 = []
  val_2 = []
  length = len(str(num))

  for i in range(100):
    n = nums[-1]
    n2 = n**2
    nums2.append(n2)
    n2 = str(n2)

    mr=n2[1:]
    mr = mr[:-1]

    if not mr:
      df = pd.DataFrame({'R(n)': nums[:-1], 'R(n)^2': nums2[:-1], 'M.R(n)^2': m_r, 'Valor 1': val_1, 'Valor 2': val_2})
      return df

    val1 = mr[:length]
    val2 = mr[-length:]



    n2 = int(n2)
    val1 = int(val1)
    val2 = int(val2)
    val_1.append(val1)
    val_2.append(val2)
    mr=int(mr)
    m_r.append(mr)


    nums.append(val_1[-1])

  df = pd.DataFrame({'R(n)': nums[:-1], 'R(n)^2': nums2, 'M.R(n)^2': m_r, 'Valor 1': val_1, 'Valor 2': val_2})

  return df
```
